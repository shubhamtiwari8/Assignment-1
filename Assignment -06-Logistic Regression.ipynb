```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# Importing the dataset
bank=pd.read_csv("C:/Users/Shubham Tiwari/Downloads/bank-full.csv",delimiter=';')
bank.head()
```

# EDA

```python
bank.info()
```

```python
# One-Hot Encoding of categrical variables
data=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
data
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
data
```

```python
data.info()
```

```python
# Custom Binary Encoding of Binary o/p variables 
data['default'] = np.where(data['default'].str.contains("yes"), 1, 0)
data['housing'] = np.where(data['housing'].str.contains("yes"), 1, 0)
data['loan'] = np.where(data['loan'].str.contains("yes"), 1, 0)
data['y'] = np.where(data['y'].str.contains("yes"), 1, 0)
data
```

```python
# Find and Replace Encoding for month categorical varaible
data['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
data=data.replace(order)
```

```python
data
```

```python
data.info()
```

# Model Building

```python
# Dividing our data into input and output variables
x=pd.concat([data.iloc[:,0:11],data.iloc[:,12:]],axis=1)
y=data.iloc[:,11]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

# Model Predictions

```python
# Predict for x dataset
pred_y=classifier.predict(x)
pred_y
```

```python
pred_y1=pd.DataFrame({'actual':y,'pred_prob':pred_y})
pred_y1
```

# Testing Model Accuracy

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,pred_y)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39107+1282)/(39107+815+4007+1282)
```

```python
# As accuracy = 0.8933, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(x)[:,1] 
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,pred_y)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
